```
import pandas as pd
import numpy as np
```

```
k=pd.read_csv("readme.txt",delimiter="\t",header=None,names=["UserID","MovieID","Rating","Datetime"])
```

```
k.head()
```

```

k["Datetime"]=pd.to_datetime(k["Datetime"],unit='s')
k.head()
```

```
k["favourables"]=k['Rating']>3
k.head()
```

```
k1=k[k["UserID"].isin(range(200))]
k1.head()
```

```
k2=k1[k1["favourables"]]
k2.head()
```

```
k3=dict((a,frozenset(v.values)) for a,v in k2.groupby("UserID")["MovieID"])
k4=k2[["MovieID","favourables"]].groupby("MovieID").sum()
k4.sort_values("favourables",ascending=False)
```
