```

signals=[]
for i in x:
  signal,k=lb.load(i)
  signals.append(signal)
print(k)  
```

```
lb.display.waveplot(signals[1],sr=22050)
```

```
 mfccs=lb.feature.mfcc(signals[1],sr=22050)
```

```
librosa.display.specshow(mfccs, sr=22050, x_axis='time')
```
