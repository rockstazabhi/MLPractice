```
import os
ad=[]
import numpy as np
```

```
a=np.load("/content/drive/My Drive/sparsh/Music_sm{2075}.npy")
b=np.load("/content/drive/My Drive/sparsh/Silence_sm{1713}.npy")
c=np.load("/content/drive/My Drive/sparsh/Speech_sm{2175}.npy")
x=np.concatenate((a,b,c),axis=0)
```

```
# x.shape
x1=x.reshape((5963,220*20))
x1.shape
```

```
a1=np.tile(np.array([1,0,0]),(2075,1))
b1=np.tile(np.array([0,1,0]),(1713,1))
c1=np.tile(np.array([0,0,1]),(2175,1))
y=np.concatenate((a1,b1,c1),axis=0)
```

```

# first neural network with keras make predictions
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
# load the dataset
# dataset = loadtxt('pima-indians-diabetes.csv', delimiter=',')
# split into input (X) and output (y) variables
# X = dataset[:,0:8]
# y = dataset[:,8]
# define the keras model
model = Sequential()
model.add(Dense(8, input_dim=4400, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()
# compile the keras model

```

```
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# summarize the first 5 cases
# for i in range(5):
# 	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```

```
# fit the keras model on the dataset
model.fit(x1, y, epochs=20, batch_size=10)
# make class predictions with the model
```

```
predictions = model.predict_classes(x1)
```
